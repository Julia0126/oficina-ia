```python
def string_to_snake_case(string):
    string = re.sub(" ","_", string)
    string = re.sub("([a-z0-9])(A-Z)", r"\1_\2", string)
    return string.lower()
```

```python
df.columns = [string_to_snake_case(col) for col in df.columns]
```
